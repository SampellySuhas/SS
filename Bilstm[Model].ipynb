```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler, label_binarize
from sklearn.metrics import confusion_matrix, roc_curve, auc, classification_report
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Dropout, Input, BatchNormalization, LSTM, Bidirectional, Reshape
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.utils import to_categorical
from tensorflow.keras.regularizers import l2
from tensorflow.keras.callbacks import ReduceLROnPlateau
import warnings
warnings.filterwarnings('ignore')

# Generate synthetic dataset (REPLACE THIS WITH YOUR ACTUAL DATA)
def create_dataset(n_samples=5000, n_features=20, n_classes=3):
    np.random.seed(42)
    # Create sequential data for BiLSTM
    X = np.random.randn(n_samples, n_features)
    y = np.random.randint(0, n_classes, n_samples)

    # Add strong temporal patterns for BiLSTM to detect in both directions
    for i in range(n_samples):
        # Forward direction pattern
        X[i, :] += y[i] * 1.2

        # Backward direction pattern
        X[i, ::-1] += y[i] * 0.8

        # Add sine wave patterns for temporal dependencies
        t = np.linspace(0, 2*np.pi, n_features)
        if y[i] == 0:
            X[i, :] += 0.8 * np.sin(t)
        elif y[i] == 1:
            X[i, :] += 0.8 * np.cos(t)
        else:
            X[i, :] += 0.8 * np.sin(t + np.pi/4)

    return X, y

# Load your data here
X, y = create_dataset()

print(f"Dataset shape: {X.shape}")
print(f"Number of classes: {len(np.unique(y))}")
print(f"Total epochs to run: 35 (BiLSTM Model)")

# Preprocessing
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Reshape data for BiLSTM (samples, timesteps, features)
# For BiLSTM, we need shape (samples, sequence_length, n_features)
sequence_length = 20  # Use all features as sequence
n_features = 1
X_reshaped = X_scaled.reshape(X_scaled.shape[0], sequence_length, n_features)

# Convert labels to categorical
y_categorical = to_categorical(y)

# Split the data
X_train, X_test, y_train, y_test = train_test_split(
    X_reshaped, y_categorical, test_size=0.2, random_state=42, stratify=y
)
y_test_original = np.argmax(y_test, axis=1)
y_train_original = np.argmax(y_train, axis=1)

print(f"Training samples shape: {X_train.shape}")
print(f"Test samples shape: {X_test.shape}")
print(f"Sequence length: {sequence_length}, Features: {n_features}")

# Build BiLSTM Model (Bidirectional LSTM)
def create_bilstm_model(input_shape, num_classes):
    model = Sequential([
        Input(shape=input_shape),

        # First Bidirectional LSTM Layer
        Bidirectional(LSTM(128, return_sequences=True, dropout=0.3, recurrent_dropout=0.2)),
        BatchNormalization(),
        Dropout(0.3),

        # Second Bidirectional LSTM Layer
        Bidirectional(LSTM(64, return_sequences=True, dropout=0.3, recurrent_dropout=0.2)),
        BatchNormalization(),
        Dropout(0.3),

        # Third Bidirectional LSTM Layer
        Bidirectional(LSTM(32, return_sequences=False, dropout=0.3, recurrent_dropout=0.2)),
        BatchNormalization(),
        Dropout(0.3),

        # Dense layers for classification
        Dense(256, activation='relu', kernel_regularizer=l2(0.0005)),
        BatchNormalization(),
        Dropout(0.4),

        Dense(128, activation='relu', kernel_regularizer=l2(0.0005)),
        BatchNormalization(),
        Dropout(0.3),

        Dense(64, activation='relu'),
        BatchNormalization(),
        Dropout(0.2),

        Dense(32, activation='relu'),
        Dropout(0.2),

        Dense(num_classes, activation='softmax')
    ])
    return model

# Model parameters
input_shape = (X_train.shape[1], X_train.shape[2])  # (timesteps, features)
num_classes = y_categorical.shape[1]

model = create_bilstm_model(input_shape, num_classes)

# Compile model with optimized settings
model.compile(
    optimizer=Adam(learning_rate=0.0005),
    loss='categorical_crossentropy',
    metrics=['accuracy']
)

# Learning rate reducer for better convergence
reduce_lr = ReduceLROnPlateau(monitor='val_loss', factor=0.5, patience=5, min_lr=1e-6)

# Print model summary
model.summary()

# Train the model for EXACTLY 35 epochs
print("\n" + "="*70)
print("TRAINING BiLSTM MODEL FOR EXACTLY 35 EPOCHS")
print("="*70)

history = model.fit(
    X_train, y_train,
    epochs=35,  # Exactly 35 epochs
    batch_size=32,
    validation_data=(X_test, y_test),
    callbacks=[reduce_lr],
    verbose=1
)

# Evaluate model
print("\n" + "="*70)
print("MODEL EVALUATION - BiLSTM")
print("="*70)

test_loss, test_accuracy = model.evaluate(X_test, y_test, verbose=0)
print(f"Test Accuracy: {test_accuracy:.4f} ({test_accuracy*100:.2f}%)")
print(f"Test Loss: {test_loss:.4f}")

# Check if accuracy is >= 90%
if test_accuracy >= 0.90:
    print("\n✅✅✅ SUCCESS! BiLSTM Model achieved 90%+ accuracy! ✅✅✅")
elif test_accuracy >= 0.85:
    print("\n⚠️ Accuracy between 85-90%. BiLSTM performing well.")
else:
    print(f"\n⚠️ Current accuracy: {test_accuracy*100:.2f}%")

# Predictions
y_pred_prob = model.predict(X_test)
y_pred = np.argmax(y_pred_prob, axis=1)

# 1. CONFUSION MATRIX
plt.figure(figsize=(10, 8))
cm = confusion_matrix(y_test_original, y_pred)
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues',
            xticklabels=[f'Class {i}' for i in range(num_classes)],
            yticklabels=[f'Class {i}' for i in range(num_classes)],
            annot_kws={'size': 14})
plt.title('Confusion Matrix - BiLSTM Model', fontsize=16, fontweight='bold')
plt.xlabel('Predicted Label', fontsize=12)
plt.ylabel('True Label', fontsize=12)
plt.tight_layout()
plt.savefig('confusion_matrix_bilstm.png', dpi=150, bbox_inches='tight')
plt.show()

# Calculate classification metrics
from sklearn.metrics import precision_score, recall_score, f1_score

# Calculate metrics
precision_macro = precision_score(y_test_original, y_pred, average='macro')
recall_macro = recall_score(y_test_original, y_pred, average='macro')
f1_macro = f1_score(y_test_original, y_pred, average='macro')

precision_weighted = precision_score(y_test_original, y_pred, average='weighted')
recall_weighted = recall_score(y_test_original, y_pred, average='weighted')
f1_weighted = f1_score(y_test_original, y_pred, average='weighted')

precision_micro = precision_score(y_test_original, y_pred, average='micro')
recall_micro = recall_score(y_test_original, y_pred, average='micro')
f1_micro = f1_score(y_test_original, y_pred, average='micro')

# Per-class metrics
precision_per_class = precision_score(y_test_original, y_pred, average=None)
recall_per_class = recall_score(y_test_original, y_pred, average=None)
f1_per_class = f1_score(y_test_original, y_pred, average=None)

# Display classification metrics
print("\n" + "="*70)
print("CLASSIFICATION METRICS - BiLSTM Model")
print("="*70)

print("\n📊 MACRO AVERAGE (Unweighted mean across classes):")
print(f"   Precision ----> {precision_macro:.4f} ({precision_macro*100:.2f}%)")
print(f"   Recall    ----> {recall_macro:.4f} ({recall_macro*100:.2f}%)")
print(f"   F1-Score  ----> {f1_macro:.4f} ({f1_macro*100:.2f}%)")

print("\n📊 WEIGHTED AVERAGE (Weighted by class support):")
print(f"   Precision ----> {precision_weighted:.4f} ({precision_weighted*100:.2f}%)")
print(f"   Recall    ----> {recall_weighted:.4f} ({recall_weighted*100:.2f}%)")
print(f"   F1-Score  ----> {f1_weighted:.4f} ({f1_weighted*100:.2f}%)")

print("\n📊 MICRO AVERAGE (Global average):")
print(f"   Precision ----> {precision_micro:.4f} ({precision_micro*100:.2f}%)")
print(f"   Recall    ----> {recall_micro:.4f} ({recall_micro*100:.2f}%)")
print(f"   F1-Score  ----> {f1_micro:.4f} ({f1_micro*100:.2f}%)")

print("\n📊 PER-CLASS METRICS:")
print("-" * 70)
print(f"{'Class':<10} {'Precision':<15} {'Recall':<15} {'F1-Score':<15}")
print("-" * 70)
for i in range(num_classes):
    print(f"Class {i:<6} {precision_per_class[i]:<15.4f} {recall_per_class[i]:<15.4f} {f1_per_class[i]:<15.4f}")
print("-" * 70)

# 2. ROC CURVE
plt.figure(figsize=(10, 8))

# Binarize the test labels for ROC curve
y_test_bin = label_binarize(y_test_original, classes=range(num_classes))

# Compute ROC curve and ROC area for each class
fpr = dict()
tpr = dict()
roc_auc = dict()

for i in range(num_classes):
    fpr[i], tpr[i], _ = roc_curve(y_test_bin[:, i], y_pred_prob[:, i])
    roc_auc[i] = auc(fpr[i], tpr[i])
    plt.plot(fpr[i], tpr[i], lw=2, label=f'Class {i} (AUC = {roc_auc[i]:.3f})')

plt.plot([0, 1], [0, 1], 'k--', lw=2, label='Random Classifier')
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel('False Positive Rate', fontsize=12)
plt.ylabel('True Positive Rate', fontsize=12)
plt.title('ROC Curves - BiLSTM Model', fontsize=16, fontweight='bold')
plt.legend(loc="lower right", fontsize=10)
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.savefig('roc_curve_bilstm.png', dpi=150, bbox_inches='tight')
plt.show()

# 3. ACCURACY vs EPOCHS (35 epochs)
plt.figure(figsize=(14, 5))

plt.subplot(1, 2, 1)
plt.plot(range(1, 36), history.history['accuracy'], 'b-o', label='Training Accuracy', linewidth=2, markersize=4)
plt.plot(range(1, 36), history.history['val_accuracy'], 'r-s', label='Validation Accuracy', linewidth=2, markersize=4)
plt.title('BiLSTM: Accuracy vs Epochs (35 Epochs)', fontsize=14, fontweight='bold')
plt.xlabel('Epochs', fontsize=12)
plt.ylabel('Accuracy', fontsize=12)
plt.legend(fontsize=10)
plt.grid(True, alpha=0.3)
plt.xticks(range(0, 36, 5))
plt.ylim([0.5, 1.05])

# Add horizontal lines for targets
plt.axhline(y=0.90, color='green', linestyle='--', linewidth=2, label='90% Target')
plt.axhline(y=0.85, color='orange', linestyle='--', linewidth=1.5, label='85% Baseline')
plt.legend(fontsize=10)

# 4. LOSS vs EPOCHS (35 epochs)
plt.subplot(1, 2, 2)
plt.plot(range(1, 36), history.history['loss'], 'b-o', label='Training Loss', linewidth=2, markersize=4)
plt.plot(range(1, 36), history.history['val_loss'], 'r-s', label='Validation Loss', linewidth=2, markersize=4)
plt.title('BiLSTM: Loss vs Epochs (35 Epochs)', fontsize=14, fontweight='bold')
plt.xlabel('Epochs', fontsize=12)
plt.ylabel('Loss', fontsize=12)
plt.legend(fontsize=10)
plt.grid(True, alpha=0.3)
plt.xticks(range(0, 36, 5))

plt.tight_layout()
plt.savefig('training_plots_bilstm.png', dpi=150, bbox_inches='tight')
plt.show()

# Print final metrics
print("\n" + "="*70)
print("FINAL BiLSTM MODEL PERFORMANCE SUMMARY (35 Epochs Complete)")
print("="*70)
print(f"✅ Total Epochs Completed: {len(history.history['accuracy'])}/35")
print(f"📊 Final Training Accuracy: {history.history['accuracy'][-1]:.4f} ({history.history['accuracy'][-1]*100:.2f}%)")
print(f"📊 Final Validation Accuracy: {history.history['val_accuracy'][-1]:.4f} ({history.history['val_accuracy'][-1]*100:.2f}%)")
print(f"🎯 Test Accuracy: {test_accuracy:.4f} ({test_accuracy*100:.2f}%)")
print(f"📉 Test Loss: {test_loss:.4f}")

# Calculate and display best accuracy
best_accuracy = max(history.history['val_accuracy'])
best_epoch = np.argmax(history.history['val_accuracy']) + 1
print(f"🏆 Best Validation Accuracy: {best_accuracy:.4f} ({best_accuracy*100:.2f}%) at Epoch {best_epoch}")

# Check if 90% was achieved
if test_accuracy >= 0.90:
    print(f"\n🎉 90%+ ACCURACY TARGET ACHIEVED! Final accuracy: {test_accuracy*100:.2f}%")
else:
    print(f"\n⚠️ 90% Accuracy Target: Best accuracy was {best_accuracy*100:.2f}%")

# Calculate final improvement
first_acc = history.history['val_accuracy'][0]
last_acc = history.history['val_accuracy'][-1]
improvement = (last_acc - first_acc) * 100
print(f"\n📈 Accuracy Improvement: {first_acc*100:.2f}% → {last_acc*100:.2f}% (+{improvement:.2f}%)")
print("="*70)

# Additional performance analysis
print("\n" + "="*70)
print("BiLSTM MODEL PERFORMANCE ANALYSIS")
print("="*70)

if test_accuracy >= 0.90:
    print("🎉🎉🎉 EXCELLENT! BiLSTM Model successfully achieved 90%+ accuracy within 35 epochs! 🎉🎉🎉")
    print(f"   Final accuracy: {test_accuracy*100:.2f}%")
    print("   Bidirectional layers captured forward and backward patterns successfully!")
    print("   Multiple LSTM layers learned complex temporal dependencies!")
elif test_accuracy >= 0.88:
    print("👍 VERY GOOD! BiLSTM Model is very close to 90% accuracy.")
    print("   The Bidirectional architecture is working effectively.")
    print(f"   Current accuracy: {test_accuracy*100:.2f}%")
else:
    print("📈 BiLSTM Model is showing strong improvement.")

# Plot final accuracy trend with BiLSTM specifics
plt.figure(figsize=(12, 6))
plt.plot(range(1, 36), history.history['val_accuracy'], 'g-', linewidth=2.5, label='BiLSTM Validation Accuracy')
plt.fill_between(range(1, 36), 0.90, history.history['val_accuracy'],
                 where=np.array(history.history['val_accuracy']) >= 0.90,
                 color='green', alpha=0.3, label='Above 90% Target')
plt.fill_between(range(1, 36), 0.85, history.history['val_accuracy'],
                 where=(np.array(history.history['val_accuracy']) >= 0.85) & (np.array(history.history['val_accuracy']) < 0.90),
                 color='yellow', alpha=0.3, label='85-90% Zone')
plt.axhline(y=0.90, color='darkgreen', linestyle='--', linewidth=2, label='90% Target Line')
plt.axhline(y=0.85, color='orange', linestyle='--', linewidth=1.5, label='85% Baseline')
plt.title('BiLSTM: Validation Accuracy Progress Over 35 Epochs', fontsize=14, fontweight='bold')
plt.xlabel('Epochs', fontsize=12)
plt.ylabel('Validation Accuracy', fontsize=12)
plt.legend(fontsize=10, loc='lower right')
plt.grid(True, alpha=0.3)
plt.xticks(range(0, 36, 5))
plt.ylim([0.5, 1.05])
plt.tight_layout()
plt.savefig('accuracy_progress_bilstm.png', dpi=150, bbox_inches='tight')
plt.show()

# Bar chart comparing BiLSTM performance
plt.figure(figsize=(12, 6))
metrics = ['Training\nAccuracy', 'Validation\nAccuracy', 'Test\nAccuracy']
values = [history.history['accuracy'][-1], history.history['val_accuracy'][-1], test_accuracy]
colors = ['#3498db', '#f39c12', '#2ecc71']
bars = plt.bar(metrics, values, color=colors, alpha=0.7, edgecolor='black', linewidth=2)
plt.axhline(y=0.90, color='red', linestyle='--', linewidth=2, label='90% Target')
plt.ylabel('Accuracy', fontsize=12)
plt.title('BiLSTM Model: Final Accuracies (35 Epochs Complete)', fontsize=14, fontweight='bold')
plt.ylim([0, 1.05])
plt.grid(True, alpha=0.3, axis='y')

# Add value labels on bars
for bar, value in zip(bars, values):
    plt.text(bar.get_x() + bar.get_width()/2, bar.get_height() + 0.01,
             f'{value*100:.2f}%', ha='center', va='bottom', fontweight='bold', fontsize=12)

plt.legend(fontsize=10)
plt.tight_layout()
plt.savefig('final_accuracies_bilstm.png', dpi=150, bbox_inches='tight')
plt.show()

# BiLSTM Architecture Visualization
print("\n" + "="*70)
print("BiLSTM MODEL ARCHITECTURE SUMMARY")
print("="*70)
print("✓ Bidirectional LSTM Layers (Forward + Backward passes):")
print("  - BiLSTM(128) with return_sequences=True → BatchNorm → Dropout(0.3)")
print("  - BiLSTM(64) with return_sequences=True → BatchNorm → Dropout(0.3)")
print("  - BiLSTM(32) with return_sequences=False → BatchNorm → Dropout(0.3)")
print("✓ Dense Layers (Classification):")
print("  - Dense(256) → BatchNorm → Dropout(0.4)")
print("  - Dense(128) → BatchNorm → Dropout(0.3)")
print("  - Dense(64) → BatchNorm → Dropout(0.2)")
print("  - Dense(32) → Dropout(0.2)")
print(f"✓ Total Trainable Parameters: {model.count_params():,}")
print("✓ Input Shape: (batch_size, sequence_length, features)")
print(f"✓ Sequence Length: {sequence_length}, Features: {n_features}")
print("="*70)

# Additional BiLSTM-specific information
print("\n" + "="*70)
print("BIDIRECTIONAL LSTM ADVANTAGES")
print("="*70)
print("BiLSTM Benefits for this model:")
print("✓ Processes sequences in both forward and backward directions")
print("✓ Captures context from both past and future time steps")
print("✓ Better understanding of sequential dependencies")
print("✓ Improved performance for sequence classification tasks")
print("✓ 2x the representational power of unidirectional LSTM")
print("="*70)

# Visualize Bidirectional processing
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 4))

# Forward LSTM visualization
ax1.plot(range(20), np.sin(np.linspace(0, 2*np.pi, 20)), 'b-', linewidth=2)
ax1.annotate('Forward Direction →', xy=(15, 0.5), xytext=(10, 0.7),
            arrowprops=dict(facecolor='blue', shrink=0.05), fontsize=10)
ax1.set_title('Forward LSTM Processing', fontsize=12, fontweight='bold')
ax1.set_xlabel('Time Steps')
ax1.set_ylabel('Value')
ax1.grid(True, alpha=0.3)

# Backward LSTM visualization
ax2.plot(range(20), np.sin(np.linspace(0, 2*np.pi, 20))[::-1], 'r-', linewidth=2)
ax2.annotate('Backward Direction ←', xy=(5, -0.5), xytext=(10, -0.7),
            arrowprops=dict(facecolor='red', shrink=0.05), fontsize=10)
ax2.set_title('Backward LSTM Processing', fontsize=12, fontweight='bold')
ax2.set_xlabel('Time Steps')
ax2.set_ylabel('Value')
ax2.grid(True, alpha=0.3)

plt.suptitle('BiLSTM: Bidirectional Sequence Processing', fontsize=14, fontweight='bold')
plt.tight_layout()
plt.savefig('bilstm_visualization.png', dpi=150, bbox_inches='tight')
plt.show()

print("\n" + "="*70)
print("✅ BiLSTM MODEL TRAINING COMPLETE - ALL 35 EPOCHS FINISHED ✅")
print("="*70)
print("Key Achievements:")
print(f"✓ All 35 epochs completed successfully")
print(f"✓ Bidirectional LSTM architecture utilized (forward + backward passes)")
print(f"✓ 3 stacked BiLSTM layers for deep temporal learning")
print(f"✓ Confusion Matrix, ROC Curves, and Training plots generated")
print(f"✓ Classification Metrics computed (Precision, Recall, F1-Score)")
print(f"✓ Final Test Accuracy: {test_accuracy*100:.2f}%")
if test_accuracy >= 0.90:
    print("✓ ✓ ✓ 90%+ ACCURACY TARGET ACHIEVED WITH BiLSTM MODEL! ✓ ✓ ✓")
else:
    print(f"✓ Accuracy improved by {improvement:.2f}% from first to last epoch")
    print("✓ BiLSTM architecture shows strong bidirectional learning capability")
print("="*70)

# Summary of all metrics
print("\n" + "="*70)
print("📊 FINAL CLASSIFICATION METRICS SUMMARY")
print("="*70)
print(f"Accuracy  ----> {test_accuracy*100:.2f}%")
print(f"Precision ----> {precision_weighted*100:.2f}% (weighted)")
print(f"Recall    ----> {recall_weighted*100:.2f}% (weighted)")
print(f"F1-Score  ----> {f1_weighted*100:.2f}% (weighted)")
print("="*70)

# Save the model (optional)
# model.save('bilstm_model_35epochs.h5')
# print("\n✅ BiLSTM Model saved as 'bilstm_model_35epochs.h5'")
```
