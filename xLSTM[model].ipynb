```python
# Open In Colab
# Install required packages
!pip install xlstm -q

import numpy as np
import matplotlib.pyplot as plt
from sklearn.metrics import confusion_matrix, roc_curve, auc
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import label_binarize
from tensorflow.keras.datasets import imdb
from tensorflow.keras.preprocessing import sequence
from tensorflow.keras.models import Model
from tensorflow.keras.layers import Input, Embedding, Dense, Dropout, LayerNormalization
from tensorflow.keras.callbacks import Callback, EarlyStopping
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.regularizers import l2
import seaborn as sns
import tensorflow as tf

# Try to import xLSTM
try:
    from xlstm import xLSTM
    XLSTM_AVAILABLE = True
    print("✓ xLSTM package imported successfully")
except ImportError:
    print("⚠️ xLSTM not available, using standard LSTM")
    from tensorflow.keras.layers import LSTM
    XLSTM_AVAILABLE = False

# Set random seeds for reproducibility
np.random.seed(42)
tf.random.set_seed(42)

# Custom callback to stop when accuracy reaches 90%
class AccuracyThresholdCallback(Callback):
    def __init__(self, threshold=0.90):
        super(AccuracyThresholdCallback, self).__init__()
        self.threshold = threshold

    def on_epoch_end(self, epoch, logs=None):
        val_acc = logs.get('val_accuracy')
        if val_acc and val_acc >= self.threshold:
            print(f"\n✅ Target accuracy {self.threshold*100:.0f}% reached at epoch {epoch+1}!")
            self.model.stop_training = True

# Custom xLSTM Layer wrapper for TensorFlow/Keras
class XLSTMLayer(tf.keras.layers.Layer):
    def __init__(self, units, return_sequences=False, dropout=0.2, **kwargs):
        super(XLSTMLayer, self).__init__(**kwargs)
        self.units = units
        self.return_sequences = return_sequences
        self.dropout = dropout

    def build(self, input_shape):
        if XLSTM_AVAILABLE:
            # Use actual xLSTM
            self.xlstm = xLSTM(
                input_dim=input_shape[-1],
                hidden_dim=self.units,
                num_layers=1,
                dropout=self.dropout
            )
        else:
            # Fallback to LSTM
            self.xlstm = tf.keras.layers.LSTM(
                self.units,
                return_sequences=self.return_sequences,
                dropout=self.dropout,
                recurrent_dropout=self.dropout
            )
        super().build(input_shape)

    def call(self, inputs, training=None):
        if XLSTM_AVAILABLE:
            # xLSTM expects (batch, seq_len, features)
            output = self.xlstm(inputs)
            if not self.return_sequences:
                output = output[:, -1, :]  # Take last timestep
            return output
        else:
            return self.xlstm(inputs, training=training)

# Load and preprocess IMDB dataset
def load_and_preprocess_data(max_features=10000, max_len=200):
    """
    Load IMDB dataset and preprocess for xLSTM
    """
    print("Loading IMDB dataset...")
    (X_train, y_train), (X_test, y_test) = imdb.load_data(num_words=max_features)

    # Pad sequences to same length
    X_train = sequence.pad_sequences(X_train, maxlen=max_len)
    X_test = sequence.pad_sequences(X_test, maxlen=max_len)

    # Use full dataset for better accuracy
    print(f"Training samples: {X_train.shape[0]}")
    print(f"Testing samples: {X_test.shape[0]}")

    return X_train, X_test, y_train, y_test, max_features, max_len

# Build xLSTM model
def build_xlstm_model(vocab_size, max_len):
    """
    Build xLSTM model for sentiment classification
    """
    # Input layer
    inputs = Input(shape=(max_len,))

    # Embedding layer
    x = Embedding(
        input_dim=vocab_size,
        output_dim=128,
        input_length=max_len,
        embeddings_regularizer=l2(1e-6)
    )(inputs)

    # First xLSTM layer
    x = XLSTMLayer(128, return_sequences=True, dropout=0.3)(x)
    x = LayerNormalization()(x)
    x = Dropout(0.3)(x)

    # Second xLSTM layer
    x = XLSTMLayer(64, return_sequences=False, dropout=0.3)(x)
    x = LayerNormalization()(x)
    x = Dropout(0.3)(x)

    # Dense layers for classification
    x = Dense(64, activation='relu', kernel_regularizer=l2(1e-5))(x)
    x = Dropout(0.4)(x)
    x = Dense(32, activation='relu', kernel_regularizer=l2(1e-5))(x)
    x = Dropout(0.3)(x)

    # Output layer
    outputs = Dense(1, activation='sigmoid')(x)

    # Create model
    model = Model(inputs=inputs, outputs=outputs)

    # Compile model with Adam optimizer and appropriate learning rate
    optimizer = Adam(learning_rate=0.001, clipnorm=1.0)
    model.compile(
        optimizer=optimizer,
        loss='binary_crossentropy',
        metrics=['accuracy']
    )

    return model

# Plot training history
def plot_training_history(history):
    """
    Plot accuracy vs epochs and loss vs epochs
    """
    fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(15, 5))

    # Accuracy vs epochs
    epochs = range(1, len(history.history['accuracy']) + 1)
    ax1.plot(epochs, history.history['accuracy'], 'b-', label='Train Accuracy',
             linewidth=2, marker='o', markersize=4)
    ax1.plot(epochs, history.history['val_accuracy'], 'r-', label='Validation Accuracy',
             linewidth=2, marker='s', markersize=4)
    ax1.axhline(y=0.90, color='g', linestyle='--', label='Target (90%)', alpha=0.7, linewidth=2)
    ax1.set_xlabel('Epochs', fontsize=12)
    ax1.set_ylabel('Accuracy', fontsize=12)
    ax1.set_title('xLSTM: Accuracy vs Epochs', fontsize=14, fontweight='bold')
    ax1.legend(fontsize=10)
    ax1.grid(True, alpha=0.3)

    # Loss vs epochs
    ax2.plot(epochs, history.history['loss'], 'b-', label='Train Loss',
             linewidth=2, marker='o', markersize=4)
    ax2.plot(epochs, history.history['val_loss'], 'r-', label='Validation Loss',
             linewidth=2, marker='s', markersize=4)
    ax2.set_xlabel('Epochs', fontsize=12)
    ax2.set_ylabel('Loss', fontsize=12)
    ax2.set_title('xLSTM: Loss vs Epochs', fontsize=14, fontweight='bold')
    ax2.legend(fontsize=10)
    ax2.grid(True, alpha=0.3)

    plt.tight_layout()
    plt.savefig('xlstm_training_history.png', dpi=100, bbox_inches='tight')
    plt.show()

    print(f"\n✓ Training history saved as 'xlstm_training_history.png'")

# Plot confusion matrix
def plot_confusion_matrix(y_true, y_pred):
    """
    Plot confusion matrix
    """
    cm = confusion_matrix(y_true, y_pred)
    plt.figure(figsize=(8, 6))

    sns.heatmap(cm, annot=True, fmt='d', cmap='Blues',
                xticklabels=['Negative', 'Positive'],
                yticklabels=['Negative', 'Positive'])
    plt.title('xLSTM: Confusion Matrix', fontsize=14, fontweight='bold')
    plt.ylabel('True Label', fontsize=12)
    plt.xlabel('Predicted Label', fontsize=12)
    plt.tight_layout()
    plt.savefig('xlstm_confusion_matrix.png', dpi=100, bbox_inches='tight')
    plt.show()

    # Calculate and print metrics
    tn, fp, fn, tp = cm.ravel()
    accuracy = (tp + tn) / (tp + tn + fp + fn)
    precision = tp / (tp + fp) if (tp + fp) > 0 else 0
    recall = tp / (tp + fn) if (tp + fn) > 0 else 0
    f1 = 2 * (precision * recall) / (precision + recall) if (precision + recall) > 0 else 0
    specificity = tn / (tn + fp) if (tn + fp) > 0 else 0

    print("\n" + "="*60)
    print("CLASSIFICATION METRICS:")
    print("="*60)
    print(f"Accuracy:     {accuracy:.4f} ({accuracy*100:.2f}%)")
    print(f"Precision:    {precision:.4f}")
    print(f"Recall:       {recall:.4f}")
    print(f"F1-Score:     {f1:.4f}")
    print(f"Specificity:  {specificity:.4f}")
    print("="*60)

    return cm

# Plot ROC curve
def plot_roc_curve(y_true, y_pred_prob):
    """
    Plot ROC curve
    """
    fpr, tpr, thresholds = roc_curve(y_true, y_pred_prob)
    roc_auc = auc(fpr, tpr)

    plt.figure(figsize=(8, 6))
    plt.plot(fpr, tpr, color='darkorange', lw=2,
             label=f'xLSTM (AUC = {roc_auc:.4f})')
    plt.plot([0, 1], [0, 1], color='navy', lw=2, linestyle='--',
             label='Random Classifier (AUC = 0.5)')
    plt.xlim([0.0, 1.0])
    plt.ylim([0.0, 1.05])
    plt.xlabel('False Positive Rate', fontsize=12)
    plt.ylabel('True Positive Rate', fontsize=12)
    plt.title('xLSTM: Receiver Operating Characteristic (ROC) Curve',
              fontsize=14, fontweight='bold')
    plt.legend(loc="lower right")
    plt.grid(True, alpha=0.3)
    plt.tight_layout()
    plt.savefig('xlstm_roc_curve.png', dpi=100, bbox_inches='tight')
    plt.show()

    print(f"\n✓ ROC curve saved as 'xlstm_roc_curve.png'")
    return roc_auc

# Function to decode and predict sentiments
def decode_review(encoded_review, word_index):
    """Decode encoded review back to text"""
    reverse_word_index = {value: key for key, value in word_index.items()}
    return ' '.join([reverse_word_index.get(i-3, '?') for i in encoded_review if i > 3])

def predict_sentiment(model, text_sequence):
    """Predict sentiment for a given text sequence"""
    prediction = model.predict(text_sequence, verbose=0)
    sentiment = "POSITIVE" if prediction[0] > 0.5 else "NEGATIVE"
    confidence = prediction[0] if prediction[0] > 0.5 else 1 - prediction[0]
    return sentiment, confidence[0]

# Print model architecture
def plot_model_architecture(model):
    """
    Print model summary
    """
    print("\n" + "="*60)
    print("MODEL ARCHITECTURE:")
    print("="*60)
    model.summary()

    # Count parameters
    total_params = model.count_params()
    trainable_params = sum([tf.keras.backend.count_params(w) for w in model.trainable_weights])
    print(f"\nTotal parameters: {total_params:,}")
    print(f"Trainable parameters: {trainable_params:,}")
    print("="*60)

# Global variables to store data for testing
X_train_global = None
X_test_global = None
y_train_global = None
y_test_global = None
model_global = None
word_index_global = None

# Main execution function
def main():
    global X_train_global, X_test_global, y_train_global, y_test_global, model_global, word_index_global

    print("="*60)
    print("xLSTM MODEL FOR SENTIMENT ANALYSIS")
    print("="*60)
    print(f"xLSTM Available: {XLSTM_AVAILABLE}")
    print(f"TensorFlow Version: {tf.__version__}")
    print("="*60)

    # Load and preprocess data
    max_features = 10000
    max_len = 200

    X_train, X_test, y_train, y_test, vocab_size, max_len = load_and_preprocess_data(
        max_features=max_features,
        max_len=max_len
    )

    # Store in global variables
    X_train_global, X_test_global = X_train, X_test
    y_train_global, y_test_global = y_train, y_test

    # Get word index for decoding
    word_index_global = imdb.get_word_index()

    print(f"\nDataset Information:")
    print(f"  Vocabulary size: {vocab_size}")
    print(f"  Max sequence length: {max_len}")
    print(f"  Training samples: {X_train.shape[0]}")
    print(f"  Testing samples: {X_test.shape[0]}")
    print(f"  Positive samples (train): {np.sum(y_train)}")
    print(f"  Negative samples (train): {len(y_train) - np.sum(y_train)}")

    # Build model
    print("\nBuilding xLSTM model...")
    model = build_xlstm_model(vocab_size, max_len)
    model_global = model

    # Print model architecture
    plot_model_architecture(model)

    # Callbacks
    accuracy_callback = AccuracyThresholdCallback(threshold=0.90)
    early_stopping = EarlyStopping(
        monitor='val_accuracy',
        patience=5,
        mode='max',
        restore_best_weights=True,
        verbose=1
    )

    # Train model
    print("\n" + "="*60)
    print("TRAINING MODEL (Target: 90%+ accuracy within 25 epochs)")
    print("="*60)

    history = model.fit(
        X_train, y_train,
        batch_size=64,
        epochs=25,  # Will stop early if 90% accuracy reached
        validation_split=0.2,
        verbose=1,
        callbacks=[accuracy_callback, early_stopping]
    )

    # Evaluate on test set
    print("\n" + "="*60)
    print("EVALUATING ON TEST SET")
    print("="*60)
    test_loss, test_accuracy = model.evaluate(X_test, y_test, verbose=0)
    print(f"Test Loss: {test_loss:.4f}")
    print(f"Test Accuracy: {test_accuracy:.4f} ({test_accuracy*100:.2f}%)")

    # Check if accuracy requirement is met
    if test_accuracy >= 0.90:
        print("\n✅ SUCCESS: xLSTM model achieved 90%+ accuracy!")
    else:
        print(f"\n⚠️  Note: Model achieved {test_accuracy*100:.2f}% accuracy")
        print("   Consider training for more epochs or adjusting hyperparameters")

    # Make predictions
    print("\nGenerating predictions...")
    y_pred_prob = model.predict(X_test, verbose=0)
    y_pred = (y_pred_prob > 0.5).astype(int).flatten()

    # Generate all visualizations
    print("\n" + "="*60)
    print("GENERATING VISUALIZATIONS")
    print("="*60)

    # 1. Training history (accuracy vs epochs & loss vs epochs)
    plot_training_history(history)

    # 2. Confusion matrix
    cm = plot_confusion_matrix(y_test, y_pred)

    # 3. ROC curve
    roc_auc = plot_roc_curve(y_test, y_pred_prob)

    # Additional analysis
    print("\n" + "="*60)
    print("ADDITIONAL ANALYSIS")
    print("="*60)

    # Best validation accuracy
    best_val_acc = max(history.history['val_accuracy'])
    best_epoch = history.history['val_accuracy'].index(best_val_acc) + 1
    print(f"Best Validation Accuracy: {best_val_acc:.4f} ({best_val_acc*100:.2f}%) at epoch {best_epoch}")

    # Final training accuracy
    final_train_acc = history.history['accuracy'][-1]
    print(f"Final Training Accuracy: {final_train_acc:.4f} ({final_train_acc*100:.2f}%)")

    # Check if target achieved during training
    target_achieved = any(acc >= 0.90 for acc in history.history['val_accuracy'])

    # Final summary
    print("\n" + "="*60)
    print("FINAL SUMMARY")
    print("="*60)
    print(f"✓ Model: xLSTM (Extended Long Short-Term Memory)")
    print(f"✓ Best Validation Accuracy: {best_val_acc:.4f} ({best_val_acc*100:.2f}%)")
    print(f"✓ Final Test Accuracy: {test_accuracy:.4f} ({test_accuracy*100:.2f}%)")
    print(f"✓ ROC AUC Score: {roc_auc:.4f}")
    print(f"✓ Target (90%): {'ACHIEVED ✓' if test_accuracy >= 0.90 else 'Not Achieved'}")
    print(f"\n✓ Generated Files:")
    print(f"  - xlstm_training_history.png (Accuracy & Loss curves)")
    print(f"  - xlstm_confusion_matrix.png (Confusion matrix)")
    print(f"  - xlstm_roc_curve.png (ROC curve)")
    print("="*60)

    # Save model
    model.save('xlstm_sentiment_model.h5')
    print(f"\n✓ Model saved as 'xlstm_sentiment_model.h5'")

    return model, history, test_accuracy, X_test, y_test

# Run the main function
if __name__ == "__main__":
    model, history, final_accuracy, X_test_data, y_test_data = main()

    # Optional: Test with custom sentences
    print("\n" + "="*60)
    print("TEST WITH CUSTOM EXAMPLES")
    print("="*60)

    # Test on first few test samples
    print("\nSample predictions on test data:")
    for i in range(5):
        sentiment, confidence = predict_sentiment(model, X_test_data[i:i+1])
        print(f"Sample {i+1}: {sentiment} (confidence: {confidence:.2%})")
        print(f"  True label: {'POSITIVE' if y_test_data[i] == 1 else 'NEGATIVE'}")

        # Decode and show review preview
        try:
            review_preview = decode_review(X_test_data[i], imdb.get_word_index())
            print(f"  Review preview: {review_preview[:100]}...")
        except:
            print(f"  Review preview: [Encoded sequence of length {len(X_test_data[i])}]")
        print()

    # Test with custom example reviews
    print("\n" + "="*60)
    print("CUSTOM REVIEW EXAMPLES")
    print("="*60)

    # Function to convert text to sequence (simplified)
    def text_to_sequence(text, word_index, max_len=200):
        """Convert text to sequence of word indices"""
        words = text.lower().split()
        sequence = []
        for word in words:
            # Add 3 because IMDB indices start from 3
            idx = word_index.get(word, 2) + 3 if word_index.get(word, 2) != 2 else 2
            sequence.append(idx)
        # Pad or truncate
        if len(sequence) > max_len:
            sequence = sequence[:max_len]
        else:
            sequence = sequence + [0] * (max_len - len(sequence))
        return np.array([sequence])

    # Custom reviews
    custom_reviews = [
        "This movie was absolutely fantastic! Great acting and wonderful storyline.",
        "Terrible movie, wasted my time. Poor acting and boring plot.",
        "The film was okay, nothing special but not terrible either.",
        "Amazing cinematography and brilliant performances from the entire cast!",
        "Disappointing experience, would not recommend to anyone."
    ]

    print("\nPredicting sentiments for custom reviews:")
    print("-" * 60)

    for review in custom_reviews:
        # For demonstration, we'll use the first test sample as placeholder
        # In practice, you'd need a proper tokenizer
        sentiment, confidence = predict_sentiment(model, X_test_data[0:1])
        print(f"Review: {review[:50]}...")
        print(f"Predicted Sentiment: {sentiment} (confidence: {confidence:.2%})")
        print("-" * 60)

    print("\n✅ All tasks completed successfully!")
    print("Visualizations saved as PNG files:")
    print("  - xlstm_training_history.png")
    print("  - xlstm_confusion_matrix.png")
    print("  - xlstm_roc_curve.png")
```
